```python
def remove_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

cleaned_text = remove_punctuation(lowercase_text)
print(cleaned_text)
```

```python
def chunk_text_by_words(text, word_count):
    words = text.split()
    chunks = []
    for i in range(0, len(words), word_count):
        chunk = ' '.join(words[i:i + word_count])
        chunks.append(chunk)
    return chunks

text_chunks = chunk_text_by_words(cleaned_text, 200)
print(f"Generated {len(text_chunks)} text chunks.")
print("First chunk:\n", text_chunks[0])
```
